# Champs de direction

Les principes et les lois qui constituent le comportement du monde naturel sont des énoncés ou des relations mettant en jeu des taux exprimant la variation ou le rapport entre deux grandeurs. En termes mathématiques, ces relations sont des équations et ces taux sont des dérivées. Les équations qui impliquent des dérivées sint appelées **équation différentielles**.

Pour répondre à cette deuxième question on peut commencer par observer l'équation d'un point de vue géométrique :

L'équation concidérée est donc $$\frac{dv}{dt} = 9,8 - \frac{v}{5}.$$


- **Point de vue géométrique :**

Supposons que $v$ a une certaine valeur ;

On peut donc, en évaluant le membre de droite de l'équation, trouver la valeur correspondante de $\frac{dv}{dt}$ ;

Par exemple si $v = 40$, alors $\frac{dv}{dt} = 1,8$, cela signifie que la pente d'une solution $v = v(t)$ a la valeur $1,8$ en tout poin où $v = 40$ ; 

On peut représenter cette information sous la forme d'un graphe dans le plan $tv$ en traçant de courts segments de droite (ou des flèches) ayant une pente de $1,8$ en différents points de la droite $v = 40$ ; 

Voici comment représenter ce "quadrillage de flèche" appelé champ de direction ou champ de pente ou champ de vecteur avec python :

```python
#commençons par doter notre python des packages utiles à cette activité :
import numpy as np
import matplotlib.pyplot as plt

#définissons la fonction 
v_prime = lambda v : 9.8 - v/5
#traçons notre repère :
#nous voulons v en fonction de t :
t = np.arange(0,11) #nous prenons des valeurs de t entre 0 et 10 avec arange de numpy

v = np.linspace(40,60,21) #nous prenons des valeurs de vitesse allant de 40 jusqu'à 60 en allant de 4 en 4

tt , vv = np.meshgrid(t, v) #utilisons meshgrid de numpy pour avoir le quadrillage voulu

def plot(x,y):
    A = plt.scatter(x, y, marker = '*', linewidth = 0.1)
    plt.xlabel("temps")
    plt.ylabel("vitesse")
    plt.title("Voilà à quoi ça resemble pour l'instant")
    return A

A = plot(tt,vv)
```

Il ne nous reste plus qu'à mettre des flèches orientées selon $\frac{dv}{dt}$, ces flèches représentent des vecteurs.

```python
dt, dv = 1, v_prime(vv) 

M = np.hypot(dt,dv)

plt.scatter(tt, vv, marker = '*', linewidth = 0.1)
plt.quiver(tt,vv ,dt/M, dv/M, scale_units='xy',scale=1.5)
plt.xlabel("temps")
plt.ylabel("vitesse")
plt.title("Voilà à quoi ça resemble avec les vecteurs")
plt.show()
```

```python
plt.quiver(tt,vv ,dt/M, dv/M, scale_units='xy',scale=1.5)
plt.xlabel("temps")
plt.ylabel("vitesse")
plt.title("Champs de direction")
plt.show()
```

**Important :** Chaque vecteur est une tangent au graphe d'une solution à l'équation.

On peut donc, même sans connaître une formule explicite des solutions, tirer des conclusions qualitatives sur le comportement des solutions ; par exemple ici, quand $v$ est inférieur à une certaine valeur alors la vitesse augmente, a contrario si $v$ est au dessus d'une certaine valeur critique la vitesse diminue. Quelle est cette valeur critique ? On peut se rapporter à l'équation et se demander à quelle valeur de $v$, sa dérivée par rapport au temps sera nulle ?

La réponse est $49m/s$.

En fait cette valeur limite est solution à l'équation, c'est une solution constante (un point d'équilibre/solution d'équilibre) :

```python
plt.quiver(tt,vv ,dt/M, dv/M, scale_units='xy',scale=1.5,label="champs de direction")
plt.axhline(y=49, xmin=0, xmax=10, linewidth = 1, color = "red", label = "valeur critique v = 49")
plt.xlabel("temps")
plt.ylabel("vitesse")
plt.legend()
plt.title("Champs de direction")
plt.show()
```

**Remarque :** La solution constante $v \equiv 49$ correspond à un moment d'équilibre entre la gravité et la traînée (pour notre exemple).

Par le graphique on peut voir que toutes les solutions semblent converger vers la solution d'équilibre quand $t$ augmente.

On peut refaire de même pour d'autres valeurs de $m$ et $\gamma$. *Question : que semble-t-il se passer ?*

## Conclusion : 
**Les champs de direction** sont des outils pratiques dans l'étude des solutions aux équations différentielles de la forme $$\frac{dy}{dt} = f(t,y),$$ où $f$ est une fonction donnée en les variables $t$ et $y$. 

Cette fonction est d'ailleurs parfois appelée la **fonction de taux.**

Dans notre exemple notre fonction $f$ est plus simple car c'est une fonction de la variable dépendante seulement. *(NB : variable dépendante car elle est dépendante du temps ; c'est une fonction).*

De manière générale on peut tracer ce champ de direction en évaluant $f$ en chaque point d'une grille rectangulaire composée d'au moins quelques centaines de points. Ensuite en chaque point de la grille, on trace un court segment de droite dont la pente est la valeur de $f$ en ce point. Ainsi chaque segment de droite est tangent au graphe de la solution passant par ce point.
