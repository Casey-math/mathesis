# Champ des tangentes et nullclines

### Nous allons prendre pour exemple l'équation : $$y' = f(t,y) = t - y^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_prime = lambda t, y : t - y**2
```

**Champ de tangentes :** A tout point $M = (t_0, y_0)$, on associe la droite $D_M$ passant par $M$ et de coefficient directeur $f(t_0,y_0)$ : $$D_M : y - y_0 = f(t_0, y_0)(t - t_0)$$
L'application $M \to D_M$ est appelée *champ des tangentes* associé à l'équation $(E) : y' = f(t,y)$.
Une courbe intégrale de $(E)$ est une courbe *différentiable* $C$ qui a pour tangente en chaque point $M \in C$ la droite $D_M$ du champ des tangentes.

```python
y = np.linspace(-1.5,1.5,7)

t = np.linspace(-1.5,1.5,11)

tt, yy = np.meshgrid(t, y)

dt, dy = 1, y_prime(tt, yy)

N = np.hypot(dt, dy)

plt.quiver(tt, yy, dt/N, dy/N)
plt.grid(color='gray', linestyle=':', linewidth=.5)
plt.show()
```

**Lignes isoclines de $(E)$ :** par définition ce sont les courbes : $$\Gamma_p : f(t,y) = p$$ correspondant à l'ensemble des points $M$ où la droite $D_M$ a une pente donnée $p$.
La courbe $\Gamma_0$ joue un rôle intéressant ; on a en effet un régionnement de $U$ : 
$$ U \cup U_+ \cup U_- \cup \Gamma_0$$ où $$U_+ = \{M \in U \mid f(M) > 0 \}, ~ U_- = \{M \in U \mid f(M) < 0 \}.$$
Les courbes intégrales sont croissantes dans $U_+$, décroissantes dans $U_-$, sationnaires (souvent extrêmales) sur $\Gamma_0$.

**Vocabulaire :** Résoudre le problème de Cauchy revient à trouver une *"courbe intégrale"* de $(E)$ passant par un point donné $(t_0, y_0)\in U, ~ U \subset \mathbb{R} \times \mathbb{R}^m$.

Dans notre exemple les **lignes isoclines** de l'équation $y' = f(t,y) = t - y^2$ sont les paraboles $ t = y^2 + p$

Pour la nullcline on a donc la parabole $t = y^2$, traçons la sur le graphique :

```python
nullcline_plus = lambda t : np.sqrt(x)
nullcline_moins = lambda t : - np.sqrt(x)


x = np.linspace(0, 1.5, 100)

plt.plot(x, nullcline_plus(x), color = 'r', label = "nullcline : $\Gamma_0$")
plt.plot(x, nullcline_moins(x), color = 'r')
plt.quiver(tt, yy, dt/N, dy/N, label = "champ de tangente")
plt.grid(color='gray', linestyle=':', linewidth=.5)
plt.legend()
plt.show()
```

On voit un régionnement de $U = ]-1,5, 1,5 [ \times ]-1,5, 1,5 [$, les pentes des vecteurs tangents sont croissants dans la parabole $\Gamma_0$ et décroissant à l'extérieur.
